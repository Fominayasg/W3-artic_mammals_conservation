To know if a species threaten status has become better or worse we need to know what the categories mean. In the next cell the categories are ordered from the most threaten to the less.

```python
threat_cat_order = ['Endangered', 
 '"Very rare and believed to be decreasing in numbers"', 
 '"Very rare but believed to be stable or increasing"',
 '"Less rare but believed to be threatened-requires watching"',
 'Vulnerable',
 'Near Threatened', 
 'Least Concern',
 'Lower Risk/least concern',
 'Lower Risk/conservation dependent', 
 'Lower Risk/near threatened', 
 'Data Deficient',
 'Insufficiently Known']
```

# Visual overview of all te IUCN

First of all I want to show how de threat has changed for each species through the years

```python
plt.figure(figsize=(5,1))
plt.xticks(rotation='vertical')
plt.subplots_adjust(bottom=0.15)
'''plt.legend(bbox_to_anchor=(2.05, 2), loc='upper left', borderaxespad=0.)'''
sns.histplot(x=iucn.species, y= iucn.year, legend =True, hue = iucn.category,  hue_order =threat_cat_order, palette = "Oranges")
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation='vertical')
plt.subplots_adjust(bottom=0.15)
'''plt.legend(bbox_to_anchor=(2.05, 2), loc='upper left', borderaxespad=0.)'''
sns.histplot(x=iucn.species, y= iucn.year, legend =False, hue = iucn.category,  hue_order =threat_cat_order, palette = "Oranges")
```
